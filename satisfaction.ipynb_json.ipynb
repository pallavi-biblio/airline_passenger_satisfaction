```python
import pandas as pd
from sklearn.feature_selection import mutual_info_classif, chi2
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np

# Load dataset
df = pd.read_csv('airline-passenger-satisfaction/airline_passenger_satisfaction.csv')  # Replace with your dataset path

# Target variable (adjust as necessary)
target_column = 'Satisfaction'

# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove target from features
if target_column in numerical_cols:
    numerical_cols.remove(target_column)
if target_column in categorical_cols:
    categorical_cols.remove(target_column)
if 'ID' in numerical_cols:
    numerical_cols.remove('ID')

# Handle missing values
# For numerical columns, fill with median
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())

# For categorical columns, fill with mode (most frequent)
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Encode the target variable if it's categorical
label_encoder = LabelEncoder()
df[target_column] = label_encoder.fit_transform(df[target_column])

# Preprocess categorical data by encoding it to numerical form
df_encoded = df.copy()
for col in categorical_cols:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])

# Standardize the numerical data (optional but good for some methods)
scaler = StandardScaler()
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

# Mutual Information for both numerical and categorical features
mutual_info = mutual_info_classif(df_encoded[numerical_cols + categorical_cols], df_encoded[target_column])
mutual_info_scores = pd.Series(mutual_info, index=numerical_cols + categorical_cols)
top_5_mutual_info = mutual_info_scores.sort_values(ascending=False).head(5)

# Chi-Squared test for categorical features
chi2_scores, _ = chi2(df_encoded[categorical_cols], df_encoded[target_column])
chi2_scores_series = pd.Series(chi2_scores, index=categorical_cols)
top_5_chi2 = chi2_scores_series.sort_values(ascending=False).head(5)

# Display the results
print("Top 5 features based on Mutual Information:")
print(top_5_mutual_info)

print("\nTop 5 features based on Chi-Squared Test:")
print(top_5_chi2)

```

**Observations from Mutual Information Results:**
<br>Flight Distance has the highest mutual information score (0.076965), indicating it shares the most information with the target variable compared to the other features. This suggests that knowing the flight distance helps in predicting the target variable better than knowing any other feature individually.

Age has the second-highest mutual information score (0.045726), showing that it has a moderate amount of information shared with the target variable.

**Observations from Chi-Squared Test Results:**
<br>Flight Distance has the highest Chi-Squared statistic (1.014693e+07), indicating a strong association with the target variable. The p-value is 0.0, which suggests this association is statistically significant.

Arrival Delay (Chi-Squared: 4.307525e+04) and Departure Delay (Chi-Squared: 3.239706e+04) also show significant associations with the target variable, as indicated by their large Chi-Squared statistics and p-values of 0.0.

Age has the lowest Chi-Squared statistic (1.646746e+04), but its p-value is still 0.0, indicating that the association is statistically significant, albeit weaker compared to the other features.

**Method Of Choice:**
<br>The choice between mutual information and the Chi-Squared test depends on the nature of the data and the goal of feature selection.

Mutual Information is a non-parametric method that captures any kind of dependency (linear or non-linear) between a feature and the target variable. It is useful when the relationships are not strictly linear or when working with categorical data.

Chi-Squared Test measures the independence between categorical variables. It is effective for identifying features that have a strong, linear relationship with the target variable and works best when both the feature and target are categorical.

**Justification for Choice:**

If the target variable and features are not necessarily categorical or if you expect non-linear relationships, Mutual Information is the better choice because it can capture a wider range of dependencies.
If the target variable and features are categorical and the relationships are expected to be linear, the Chi-Squared Test would be more appropriate.
In this case, since the mutual information scores and Chi-Squared statistics both point to Flight Distance as the most important feature but highlight different aspects of the data relationships, using Mutual Information might be preferred if the data has non-linear relationships or mixed types (continuous and categorical).

## 6.1 ML technique 1 + Justification

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

# Load the dataset
data = df #pd.read_csv("airline-passenger-satisfaction/airline_passenger_satisfaction.csv")

# Handle missing values
# You can replace missing numerical data with the mean or median
imputer = SimpleImputer(strategy='mean')
data['Arrival Delay'] = imputer.fit_transform(data[['Arrival Delay']])

# Encode categorical features using LabelEncoder
label_encoder = LabelEncoder()
for col in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Satisfaction']:
    data[col] = label_encoder.fit_transform(data[col])

# Features (X) and target (y)
X = data.drop(columns=['Satisfaction', 'ID'])  # Drop 'ID' and target column
y = data['Satisfaction']  # Target column (Satisfaction)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(report)


```

Justification for using Decision Tree:
Interpretability: Decision trees are easily interpretable, allowing airlines to see which features (like delays or in-flight service) are most critical in predicting satisfaction.
No Assumptions: Decision trees are non-parametric, meaning they make no assumptions about the underlying data distribution.
Handling Complex Data: They can handle both numerical and categorical data efficiently, making them well-suited for datasets with mixed feature types (e.g., delays, class, seat comfort).

## 6.2 ML technique 2 + Justification

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("airline-passenger-satisfaction/airline_passenger_satisfaction.csv")

# Separate categorical and numerical columns
categorical_cols = data.select_dtypes(include=['object']).columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Handle missing values for numerical columns
imputer = SimpleImputer(strategy='mean')
data[numerical_cols] = imputer.fit_transform(data[numerical_cols])

# Encode categorical columns using LabelEncoder
label_encoder = LabelEncoder()
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

# Standardize the numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[numerical_cols])

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original dataset
data['Cluster'] = clusters

# Visualize the clusters using the first two numerical features (adjust as needed)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=scaled_data[:, 0], y=scaled_data[:, 1], hue=clusters, palette='viridis', s=100)
plt.title('K-Means Clustering Visualization')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Optionally, check the cluster centers
print("Cluster Centers:\n", kmeans.cluster_centers_)
```

K-Means clustering was chosen for the following reasons:

1. Simplicity and Efficiency:
K-Means is computationally efficient for large datasets, like the one you're working with. It works by minimizing the distance between points and their cluster centroids, making it straightforward to interpret.
2. Scalability:
It scales well with larger datasets, such as airline passenger data with multiple features (e.g., delays, seat comfort, in-flight services). While my initial attempt used the full dataset, K-Means can handle both sampled and full data efficiently after optimization.
3. Continuous Numerical Data:
K-Means works well with continuous numerical features, such as the departure delay, arrival delay, and other ratings (seat comfort, service quality). These are the types of variables that lend themselves well to the K-Means approach because it relies on calculating the mean of the data points in a cluster.
4. Segmentation for Business Insights:
Airlines aim to group passengers based on similar experiences or preferences. K-Means provides well-defined, non-overlapping clusters that help in creating customer personas or segments that can be targeted with personalized services (e.g., handling passengers who are more sensitive to delays differently from those who prioritize in-flight amenities).
5. Interpretability:
The centroids in K-Means clustering represent the "typical" passenger profile within each cluster. This makes it easier for airlines to interpret and act on the results, e.g., identifying a group that values fast boarding and punctuality versus another that prioritizes comfort.

Compare the performance of the ML techniques used.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Load the dataset
data = pd.read_csv('airline-passenger-satisfaction/airline_passenger_satisfaction.csv')

# Data Preprocessing: Handling missing values
imputer = SimpleImputer(strategy='mean')
data['Arrival Delay'] = imputer.fit_transform(data[['Arrival Delay']])

# Encode categorical features
label_encoder = LabelEncoder()
for col in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Satisfaction']:
    data[col] = label_encoder.fit_transform(data[col])

# Features (X) and target (y)
X = data.drop(columns=['Satisfaction', 'ID'])
y = data['Satisfaction']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Models to compare
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

# Initialize results dictionary
results = {}

# Train, predict and evaluate each model
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Collect metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)

    # Store results
    results[name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "AUC": auc
    }

    # Confusion Matrix and Classification Report
    print(f"\n{name} - Classification Report:")
    print(classification_report(y_test, y_pred))

    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# Convert results to DataFrame
results_df = pd.DataFrame(results).T

# Plot bar chart for performance comparison
results_df.plot(kind='bar', figsize=(12, 8), colormap='Set3')
plt.title('Performance Comparison of ML Models')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()

# Optional: ROC Curve for AUC comparison
plt.figure(figsize=(10, 6))
for name, model in models.items():
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc_score(y_test, y_pred_prob):.2f})')

plt.plot([0, 1], [0, 1], color='black', linestyle='--')
plt.title('ROC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
